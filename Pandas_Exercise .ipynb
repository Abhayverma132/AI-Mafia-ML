## Group By

Any groupby operation involves one of the following operations on the original object. They are âˆ’

- Splitting the Object

- Applying a function

- Combining the results

In many situations, we split the data into sets and we apply some functionality on each subset. In the apply functionality, we can perform the following operations âˆ’

- **Aggregation** âˆ’ computing a summary statistic

- **Transformation** âˆ’ perform some group-specific operation

- **Filtration** âˆ’ discarding the data with some condition

````python
# group imdb_df by movie genres
by_genres=imdb.groupby('genre')
by_genres
````

````python
by_genres.count()
````

````python
# get crime movies group
by_genres.get_group('Crime')
````

````python
# get mean of movie durations for each group
d_mean=imdb.groupby(['genre']).mean()
d_mean['duration']
````

````python
# change duration of all movies in a particular genre to mean duration of the group
crime=imdb.groupby('genre').get_group('Crime')
crime['duration']=crime['duration'].mean()
crime[['title','duration']]
````

````python
# drop groups/genres that do not have average movie duration greater than 120.
y=imdb.drop(imdb[imdb.duration<120].index,inplace=True)
print(y)
imdb
````

````python
# group weekday wise bikers count
grp=bikes.groupby('weekday').sum()
````
